## Resizing and Scaling

```python
def preinit(file_path):
    #load image file path 
    bit_pic = tf.io.read_file(file_path) #read file contents
    #load image and decode jpeg
    pic = tf.io.decode_jpeg(bit_pic)
    #resize image to 100x100 pixels from 3 channels
    pic = tf.image.resize(pic,(100,100))
    # divide by 255 to scale be between 0 and 1
    pic = pic / 255.0
    #returns image
    return pic
```

## Create Labelled Paired Dataset

```python
# create a set of twin image samples through a function that passes two images at a time
#(anchor, positive) => 1,1,1,1,1
# (anchor, negative) => 0,0,0,0,0

#zip dataset and loop through anchor and positive images and create new directory that outputs labelled positives as ones(1)
positives = tf.data.Dataset.zip((anchor, positive,tf.data.Dataset.from_tensor_slices(tf.ones(len(anchor)))))

#zip dataset and loop through anchor and negative images and create new directory that outputs labelled negatives as zeros(0)
negatives = tf.data.Dataset.zip((anchor, negative,tf.data.Dataset.from_tensor_slices(tf.zeros(len(anchor)))))

#join positives and negatives togethr to form a single dataset
data = positives.concatenate(negatives)
```

# MODEL ENGINEERING

## Build Embedding Layer

```python
#import dependencies

import matplotlib.pyplot as plt
import cv2
import numpy as np
from IPython.display import display, Image
import ipywidgets as widgets
import threading
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Layer, Flatten, Dense,Conv2D, Input, MaxPooling2D




#creating nueral network core block
#taking a look at journal diagram
#convolutions will be created

def build_embed():
    inp = Input(shape=(100,100,3), name = 'Input_Image')
    
    #first block
    #creating  a convolution 
    c1 = Conv2D (64, (10,10), activation='relu')(inp)
    #creating maxpooling layer 
    m1 = MaxPooling2D(64,(2,2), padding='same')(c1)
    
    #second block
    #creating  a convolution 
    c2 = Conv2D (128, (7,7), activation='relu')(m1)
    #creating maxpooling layer 
    m2 = MaxPooling2D(64,(2,2), padding='same')(c2)
    
    #third block
    c3 = Conv2D (128, (4,4), activation='relu')(m2)
    m3 = MaxPooling2D(64,(2,2), padding='same')(c3)
    
    #final embedding block
    c4 = Conv2D (256, (4,4), activation='relu')(m3)
    f1 = Flatten()(c4)
    d1 = Dense(4096, activation ='sigmoid')(f1)

    

    # output feature vector of 4096  output from each twin dataset stream
    #compile model
    return Model(inputs=[inp], outputs=[d1], name='embedding')
```

```python
embedding = build_embed()
```

```python
embedding.summary()
```

## Build Distance Layer

```python
#creating distance layer class
# subracting twin datasets from each other to determine level similarity
# output= anchor - (positive or negative )

class L1Dist(Layer):
    #using python inheritance method- Init
    
    def __init__(self, **kwargs):
        super().__init__()
        
    # similarity calculation  
    #imput embedding = anchor
    #validation_embedding = positive or negative datasets
    
    def call(self,input_embedding,validation_embedding):
        return tf.math.abs(input_embedding - validation_embedding)
```

```python
def make_siamese_model():
    #handle 2 inputs
    
    #anchor image input in the neural network
    input_image = Input(name='input_img', shape=(100,100,3))
    
    #validation_image in the neaural network
    validation_image = Input(name = 'validation_img', shape= (100,100,3))
    
    #combine with siamese distance components
    siamese_layer = L1Dist()
    siamese_layer._name = 'distance'
    
    #passing inputs through distance layer to output 4096 feature vector
    distances = siamese_layer(embedding(input_image), embedding(validation_image))
    
    #Classification layer
    classifier = Dense (1, activation='sigmoid')(distances)
    
    return Model(inputs = [input_image,validation_image],outputs = classifier, name='SiameseNetwork')
```

```python
siamese_model = make_siamese_model()
```

# Model Training

## Loss and Optimizer Setup

```python
binary_cross_loss = tf.losses.BinaryCrossentropy()
```

```python
opt = tf.keras.optimizers.Adam(1e-4) #learning right 0.0001
```

## Checkpoints Setup

```python
#setup checkpoint directory to save checkpoints
checkpoint_dir = './training_checkpoints'
#setup checkpoint prefix
checkpoint_prefix = os.path.join(checkpoint_dir, 'ckpt')
#setup checkpooint class
checkpoint = tf.train.Checkpoint(opt = opt, siamese_model = siamese_model)
```

## Train Step Function Setup

```python
# wrapping train_step function inside tf.function decorator which compiles the function into acallable tensorflow graph
@tf.function
#
def train_step(batch):
    #captures gradients from nueral network
    with tf.GradientTape() as tape:
        
        # Get anchor and positive/negative image
        X = batch[:2]
        # Get label
        y = batch[2]
        
        #forward Pass
        #its important to set training = true
        yhat = siamese_model(X, training =True)
        #calculate loss
        loss = binary_cross_loss(y,yhat)
    print
        
    #calculate gradients
    grad = tape.gradient(loss, siamese_model.trainable_variables)
    
    #calculate updated weights and apply to siamese model
    opt.apply_gradients(zip(grad, siamese_model.trainable_variables))
    
     
    #return loss
    return loss
```

## Build Train Loop

```python
# Import metric calculations
from tensorflow.keras.metrics import Precision, Recall

def train(data, EPOCHS):
    #loop through epochs
    for epoch in range(1, EPOCHS +1):
        print('\n Epoch {}/{}'.format(epoch,EPOCHS))
        progbar = tf.keras.utils.Progbar(len(data))
        
         # Creating a metric object 
        r = Recall()
        p = Precision()
        
        #loop through each batch
        for idx, batch in enumerate(data):
            #run train step here
            loss = train_step(batch)
            yhat = siamese_model.predict(batch[:2])
            r.update_state(batch[2], yhat)
            p.update_state(batch[2], yhat) 
            progbar.update(idx+1)
        print(loss.numpy(), r.result().numpy(), p.result().numpy())
        
            
        #save checkpoints
        if epoch % 10 ==0:
            checkpoint.save(file_prefix=checkpoint_prefix)
            
```

# Real Time Test

## Verification Function

```python
def verify(frame,model, detection_threshold,verification_threshold):
    # Build results array
    results = []
    #looping through and listing every image in verification image folder
    for image in os.listdir(os.path.join('application_data','verification_images')):
        
        #Grabbing input image from webcam, using prepocessing function created earlier 
        #to scale and resize imge and saving image in input_image folder
        input_img = preinit(os.path.join('application_data','input_image','input_image.jpg'))
        
        #Looping through validation images and passing through preprocess function, Why?
        validation_img = preinit(os.path.join('application_data','verification_images',image))
        
        #Make Predictions
        #using model.predict method created earlier
        result = model.predict(list(np.expand_dims([input_img, validation_img], axis =1)))
        #creating a big array of rseults
        results.append(result)

     #Detection threshold is metric above which a prediction is considered positive  
    detection = np.sum(np.array(results) > detection_threshold)
    
    #verification Threshold: Proportion of positive predictions / total positive samples( 50 selected images)
    verification = detection / len(os.listdir(os.path.join('application_data','verification_images')))# dividing by 50 images
    verified = verification > verification_threshold
    
    return results, verified
    
        
```
